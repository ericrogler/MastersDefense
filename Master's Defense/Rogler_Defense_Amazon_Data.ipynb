## Problem 1 - Production/Factory

Your company wants you to decide on which product they should focus on and how much of that product to produce. Another department at your company already acquired preliminary data and made it available for you, but you find that it hasn't been preprocessed down for immediate analysis. Additionally, your boss has an idea of the current demand for the product you choose.

In this position, I would need to:

1) Acquire the dataset and process it

2) Apply MCDM to decide between alternatives

3) Apply APIC to determine how much of a product I should create; over a timeframe if not immediate.

```python
import pandas as pd
#Remove 'nrows' to read in the entire file.
#Use t2.large for the entire dataset. Yet to test smaller version with micro currently.
data_df = pd.read_csv('data_defense/Reviews.csv', error_bad_lines=False)
data_df
```

```python
data_df.info()
```

```python
#Count number of unique products in sample.
count_df = data_df.ProductId.value_counts()
count_df = count_df.to_frame('Review Count').reset_index()
count_df.index.rename('ProductId', inplace=False)
count_df
#count_df = data_df.groupby('ProductId')['Id'].nunique()
#count_df.head()
```

```python
#Verifies that the count_df is equal to the same length as the data table it was counting.
count_df.sum()
```

```python
test_df = data_df
test_df
```

```python
#Preprocess text column to get rid of unneccessary characters

#Removes any non-english characters present.
test_df['Text'].replace(regex=True,inplace=True,to_replace=r'[^\x00-\x7F]+',value=r' ')

#Remove useless numbers and alphanumerical words; forces alphabetic characters.
test_df['Text'].replace(regex=True,inplace=True,to_replace=r'[^a-zA-Z0-9]',value=r' ')
test_df.head()
```

```python
#Sentiment Analysis of Text
#Using Textblob.
#Source: https://textblob.readthedocs.io/en/dev/
#Sentiment(polarity, subjectivity)
#Polarity is whether positive or negative. Subjectivity is how  subjective the statement is.
from textblob import TextBlob

def sentiment_calc(Text):
    try:
        return TextBlob(Text).sentiment
    except:
        return None

test_df['sentiment'] = test_df['Text'].apply(sentiment_calc)
test_df
```

```python
test_df['sentiment'] = test_df['sentiment'].astype(str)
test_df.info()
```

```python
test_df[['Polarity','Subjectivity']] = test_df['sentiment'].str.split(',',expand=True)
test_df
```

```python
#Regex Expression
#[^\d.-]+
#Match a single character not present in the list below [^\d.-]+
#^ matches any character  expect those in the range "\d.-"
#+ Quantifier — Matches between one and unlimited times, as many times as possible, giving back as needed (greedy)
#\d matches a digit (equal to [0-9])
#.- matches a single character in the list .- (case sensitive)
test_df['Polarity'].replace(regex=True,inplace=True,to_replace=r'[^\d.-]+',value=r'')
test_df['Subjectivity'].replace(regex=True,inplace=True,to_replace=r'[^\d.-]+',value=r'')
test_df
```

```python
#Converts polarity and subjectivity into int64 values
import numpy as np
test_df['Polarity'] = pd.to_numeric(test_df['Polarity'], errors='coerce')
test_df['Subjectivity'] = pd.to_numeric(test_df['Subjectivity'], errors='coerce')
test_df.info()
```

```python
#Treat this score as an "accuracy" of the report. Used to sort.
test_df['Helpfulness'] = np.where(test_df['HelpfulnessDenominator'] < 1, test_df['HelpfulnessDenominator'], test_df['HelpfulnessNumerator']/test_df['HelpfulnessDenominator'])
test_df
```

```python
#Polarity Table
polar = test_df[['ProductId', 'Polarity']].copy()
#Subjectivity Table
sub = test_df[['ProductId', 'Subjectivity']].copy()
#Helpfulness Table
helpscore = test_df[['ProductId', 'Helpfulness']].copy()
#Score Table
score = test_df[['ProductId', 'Score']].copy()
```

```python
#Take Average polarity of items based on value in column.
polar_avg = polar.groupby('ProductId', as_index=False)['Polarity'].mean()
polar_avg
```

```python
#Take Average subjectivity of items based on value in column.
sub_avg = sub.groupby('ProductId', as_index=False)['Subjectivity'].mean()
sub_avg
```

```python
#Take Average helpfulness of items based on value in column.
helpscore_avg = helpscore.groupby('ProductId', as_index=False)['Helpfulness'].mean()
helpscore_avg
```

```python
#Take Average score of items based on value in column.
score_avg = score.groupby('ProductId', as_index=False)['Score'].mean()
score_avg
```

```python
#A single command block here.
#Merge together all of the different tables, using a left join on ProductId between the tables.
#Start with Polarity
table_1 = polar_avg.merge(count_df,how='left', left_on='ProductId', right_on='index')
table_1['Review Count']=table_1['Review Count'].fillna(0)
del table_1['index']
table_1
```

```python
#Combines Subjectivity
table_2 = table_1.merge(sub_avg,how='left', left_on='ProductId', right_on='ProductId')
table_2['Subjectivity']=table_2['Subjectivity'].fillna(0)
table_2
```

```python
#Combines Helpfulness
table_3 = table_2.merge(helpscore_avg,how='left', left_on='ProductId', right_on='ProductId')
table_3['Helpfulness'] = table_3['Helpfulness'].fillna(0)
table_3
```

```python
#Combines Rating/Score
table_4 = table_3.merge(score_avg,how='left', left_on='ProductId', right_on='ProductId')
table_4['Score'] = table_4['Score'].fillna(0)
table_4
```

```python
#Reorder columns to have review count show up after ProductId
table_edit = table_4[['ProductId','Review Count','Helpfulness','Polarity','Subjectivity','Score']]
table_edit
```

```python
#Now to sort alternatives based on "accuracy"
#This value is done through sorting Review Count.
#The value to the right of Review Count is Helpfulness, which is an accuracy indicator.
#Helpfulness is simultaneously one of our criteria we will use for MCDM.
sorted_table = table_edit.sort_values(["Review Count","Helpfulness"], ascending=False)
sorted_table = sorted_table.rename(columns = {'Review Count':'ReviewCount'})
sorted_table
```

```python
#Eliminate rows which can be 'inaccurate'
#Review count = n = sample size
sorted_table1 = sorted_table[sorted_table.ReviewCount >= 150]
#Helpfulness = peer review accuracy of whether or not reviews were useful.
sorted_table2 = sorted_table1[sorted_table1.Helpfulness >= 0.50]
sorted_table2
```

```python
#Export top 10 results out to a .csv file out to a csv for excel-based processing.

xlsx_table = sorted_table2.head(8)
xlsx_table
```

```python
#Write the file to an Excel workbook.
writer = pd.ExcelWriter('data_defense/AmazonAnalysis.xlsx')
xlsx_table.to_excel(writer,'Data')
writer.save()
```
